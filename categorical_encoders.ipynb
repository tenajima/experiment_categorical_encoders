```python
import category_encoders as ce
import numpy as np
import pandas as pd
```

```python
train = pd.DataFrame(
    {"name": ["hoge", "hoge", "fuga", "piyo", pd.NA], "target": [0, 1, 1, 0, 0]}
).convert_dtypes()
train['id'] = train['name']
train = train.set_index('id')
train
```

```python
train.dtypes
```

## BackwardDifferenceEncoder

```python
encoder = ce.BackwardDifferenceEncoder(cols=["name"])
encoder.fit_transform(train['name'])
```

## BaseNEncoder

```python
encoder = ce.BaseNEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

## CatBoostEncoder

```python
encoder = ce.CatBoostEncoder(cols=["name"])
encoder.fit(train[["name"]].convert_dtypes(), train["target"])
encoder.transform(train[["name"]])
```

複数あるとどうなるだろう?

```python
tmp = train.copy()
tmp["group"] = pd.Series(["a", "b", "c", "a", "b"], dtype="str")
tmp
```

```python
encoder = ce.CatBoostEncoder(cols=["name", "group"])
encoder.fit(tmp[["name", "group"]].convert_dtypes(), tmp["target"])
encoder.transform(tmp[["name", "group"]])
```

## BinaryEncoder

```python
encoder = ce.BinaryEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

## Hashing

```python
encoder = ce.HashingEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

n_componentsで変えられる

```python
encoder = ce.HashingEncoder(cols=["name"], n_components=5)
encoder.fit_transform(train["name"])
```

## Helmert Coding

```python
encoder = ce.HelmertEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

## JamesSteinEncoder

```python
encoder = ce.JamesSteinEncoder(cols=["name"])
encoder.fit(train[["name"]].convert_dtypes(), train["target"])
encoder.transform(train[["name"]])
```

```python
encoder = ce.JamesSteinEncoder(cols=["name", "group"])
encoder.fit(tmp[["name", "group"]].convert_dtypes(), tmp["target"])
encoder.transform(tmp[["name", "group"]])
```

## LeaveOneOut

```python
encoder = ce.LeaveOneOutEncoder(cols=["name"])
encoder.fit(train[["name"]].convert_dtypes(), train["target"])
encoder.transform(train[["name"]])
```

## M-estimate

```python
encoder = ce.MEstimateEncoder(cols=["name"])
encoder.fit(train["name"], train["target"])
encoder.transform(train["name"])
```

## One Hot

```python
encoder = ce.OneHotEncoder(cols=["name"], use_cat_names=True)
encoder.fit_transform(train["name"])
```

## Ordinal

```python
encoder = ce.OrdinalEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

## Polynomial

```python
encoder = ce.PolynomialEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

## Sum Coding

```python
encoder = ce.SumEncoder(cols=["name"])
encoder.fit_transform(train["name"])
```

```python
encoder = ce.WOEEncoder(cols=["name"])
encoder.fit(train["name"], train['target'])
encoder.transform(train['name'])
```
